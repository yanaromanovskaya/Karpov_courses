```python
df=pd.read_csv('delays (1).csv')
df.head()
```

```python
df.delay=pd.to_timedelta(df.delay.str.replace('-',''))
```

#### Разбейте время на интервалы и поместите соответствующие значения в колонку delay_categorical , чтобы значения были разбиты по категориям : less than 1 day , 1-2 days ,2-3 days, more than 3 days
        

```python
bins=[pd.to_timedelta('0d'),
     pd.to_timedelta('1d'),
     pd.to_timedelta('2d'),
     pd.to_timedelta('3d'),
     pd.to_timedelta('100d')]
```

```python
labels=['less than 1 day',
        '1-2 days',
        '2-3 days',
        'more than 3 days']
```

```python
df['delay_categorical']=pd.cut(df.delay,
                               bins=bins,
                              labels=labels)
```
